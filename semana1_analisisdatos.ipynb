```python
import pandas as pd
import numpy as np    #llamamos las lobririas panda y numpy
import io
```

```python
from google.colab import files # cargamos el archivo de la base datos ibm a colab
uploaded = files.upload()
```

```python
df=pd.read_csv(io.StringIO(uploaded['auto.csv'].decode('utf-8'))) # traslador la base de datos a la funcion df y mostramos las 5 primeras filas con la funcio head
df.head()
```

```python
print("The last 10 rows of the dataframe\n")
df.tail(10)#  esta funcion devuelve las ultimas 10 filas del dataframe
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers) #creamos unos nombres para el encabezado del dataframe.
```

```python
df.columns = headers
df.head(10)  # reemplaza el nombre de los encabezados de las columnas
```

```python
df1=df.replace('?',np.NaN)# se reemplaza las celdas que contengan el simbolo "?" por la palabra Nan.
df.head(20)
```

```python
df=df1.dropna(subset=["price"], axis=0) # se elimina los valores faltantes de la columna
df.head(20)
```

```python
print(df.columns)# imprime el nombre de las columnas
```

```python
df.describe()  # para  obtener un resumen estadístico de cada columna, p. recuento, valor medio de columna, desviación estándar de columna, etc., usamos el método de descripción
```
